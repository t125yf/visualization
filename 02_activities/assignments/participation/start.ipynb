```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50)
 #scatterplot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)

fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)


ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')

font1 = {'family':'sans-serif'
,
'color':'blue'
,
'size':20}
font2 = {'family':'monospace'
,
'color':'green'
,
'size':14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time'
, fontdict = font1)
ax.set_ylabel('Total growth'
, fontdict = font2)
ax.set_xlabel('Years since start'
, fontdict = font2)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time'
, fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth'
, fontdict = font2)
ax.set_xlabel('Years since start'
, fontdict = font2)
```
