```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load the data
df = pd.read_excel("wellbeing-toronto-education.xlsx", sheet_name="RawData_Ref2008")

# Define predictor and response variables
X = df[["Catholic University Applicants"]]
y = df["Catholic School Literacy"]

# Initialize and fit the regression model
model = LinearRegression()
model.fit(X, y)

# Predict literacy values
df["Predicted Literacy"] = model.predict(X)

# Display model coefficients
print(f"Intercept: {model.intercept_:.4f}")
print(f"Coefficient: {model.coef_[0]:.4f}")
print(f"R² score: {r2_score(y, df['Predicted Literacy']):.4f}")

# Create scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Catholic University Applicants", y="Catholic School Literacy", label="Actual", s=60)
sns.lineplot(data=df, x="Catholic University Applicants", y="Predicted Literacy", label="Regression Line", color='green')
plt.title("Dependency: University Applicants vs. Catholic School Literacy")
plt.xlabel("Catholic University Applicants")
plt.ylabel("Catholic School Literacy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
